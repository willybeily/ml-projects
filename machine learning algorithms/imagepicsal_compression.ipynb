```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
img = Image.open("strange.0.jpg")
```

```python
np_img = np.array(img)
```

```python
plt.imshow(np_img)
```

```python
np_img.dtype
```

```python
np_img.shape
```

```python
X = np_img.reshape(-1, 3)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=8)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
pallets = pallets[np.random.randint(0, 8, 8)]
```

```python
pallets
```

```python
plt.imshow(pallets.reshape(2, 4, 3))
```

```python
output = np.array([pallets[label] for label in model.labels_])
```

```python
output.shape
```

```python
out_img = output.reshape(np_img.shape)
```

```python
plt.imshow(out_img)
```
